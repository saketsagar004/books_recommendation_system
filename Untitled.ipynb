# popularity based recommender system

```python
ratings
```

```python
books
```

```python
books_ratings = ratings.merge(books,on = "ISBN")
```

```python
books_ratings
```

```python
books_numrating_df = books_ratings.groupby("Book-Title").count()[["Book-Rating"]].reset_index()
books_numrating_df
```

```python
books_numrating_df.rename(columns = {"Book-Rating":"Num_rating"},inplace = True)
books_numrating_df
```

```python
# books_avgrating_df
books_avgrating_df = books_ratings.groupby("Book-Title").mean()[["Book-Rating"]].reset_index()
books_avgrating_df.rename(columns = {"Book-Rating":"avg_rating"},inplace = True)
books_avgrating_df
```

```python
popularity = books_numrating_df.merge(books_avgrating_df,on = "Book-Title")
popularity
```

```python
popularity_50 = popularity[popularity["Num_rating"] >=250].sort_values("avg_rating",ascending=False)
```

```python
popularity_50
```

```python
popularity_50.head(50)
```

```python
popular_df = popularity_50.head(50)
```

```python
popular_df = popular_df.merge(books,on = "Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M","Year-Of-Publication","Num_rating","avg_rating"]]
```

# collaborative filtering

```python
books_ratings
```

```python
books_ratings.groupby("User-ID").count()["Book-Rating"] > 200
```

```python
users_250 = books_ratings.groupby("User-ID").count()["Book-Rating"] > 200
book_readers = users_250[users_250].index
```

```python
filter_rating = books_ratings[books_ratings["User-ID"].isin(book_readers)]
filter_rating
```

```python
filter_rating_50 = filter_rating.head(50)
```

```python
filter_rating_50
```

```python
filter_rating.groupby("Book-Title").count()["Book-Rating"] >= 50
```

```python
book_50 = filter_rating.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_books = book_50[book_50].index
```

```python
final_ratings = filter_rating[filter_rating["Book-Title"].isin(famous_books)]
final_ratings.drop_duplicates("Book-Title")
```

```python
users_books = final_ratings.pivot_table(index = "Book-Title",columns = "User-ID",values = "Book-Rating")
users_books
```

```python
users_books.fillna(0,inplace = True)
```

```python
users_books
```

```python
users_books.index
```

```python
book_names= list(users_books.index.values)
```

```python
book_names
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(users_books)
```

```python
similarity_scores
```

```python
similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
# creating a funtion recommend
def recommend(book_name):
    index = np.where(users_books.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:8]
    data = []
    for book in similar_books:
        item = []
        temp_df = books[books["Book-Title"] == users_books.index[book[0]]]

        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))

        data.append(item)
    return data
```

```python
recommend("4 Blondes")
```
